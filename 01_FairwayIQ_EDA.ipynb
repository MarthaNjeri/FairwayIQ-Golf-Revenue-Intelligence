```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

members = pd.read_csv("output/members.csv")
revenue = pd.read_csv("output/revenue.csv")
engagement = pd.read_csv("output/engagement.csv")
status = pd.read_csv("output/member_status.csv")

print("Revenue loaded:", revenue.shape)
```

```python
revenue["TotalSpend"] = (
    revenue["MembershipFee"] +
    revenue["ProShopSpend"] +
    revenue["F&BSpend"] +
    revenue["CoachingRevenue"]
)
```

```python
## Revenue Contribution by Membership Category

Â£We evaluate which membership tier drives the highest value for the club.
# Merge member type into revenue
rev_membership = revenue.merge(members[["MemberID", "MembershipType"]], on="MemberID")

# Revenue by membership type
rev_type = rev_membership.groupby("MembershipType")["TotalSpend"].sum().sort_values(ascending=False)

print("Revenue by Membership Type:")
display(rev_type)

# Plot
rev_type.plot(kind="bar")
plt.title("Total Revenue by Membership Type")
plt.xlabel("Membership Type")
plt.ylabel("Total Revenue")
plt.show()
```
